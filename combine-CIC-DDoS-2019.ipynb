```python
import pandas as pd
import glob

files = glob.glob("/Users/nehahooda/Desktop/Intrusion-Detection/CIC-DDoS2019-cleaning/*.parquet")
print("Found parquet files:", len(files))

dfs = [pd.read_parquet(f) for f in files]
df = pd.concat(dfs, ignore_index=True)

print("Combined shape:", df.shape)
print(df.head())

df.to_csv("CICDDoS2019-combined.csv", index=False)
print("Saved CICDDoS2019-combined.csv")
```
